```python
df=df.drop(columns=['license','house_rules','reviews per month','last review'],axis=1)
```

```python
df
```

```python
df.columns
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df['price']=df['price'].str.replace('$',"")
df['price']=df['price'].str.replace(",","")
df['price']=pd.to_numeric(df['price'])
```

```python
df
```

```python
df_price=df.groupby(by='neighbourhood')['price'].mean().sort_values(ascending=False).reset_index()
df_price.head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.barplot(x=df_price['neighbourhood'].head(10),y=df_price['price'].head(10))
plt.title("Top 10 most expensive neighbourhood")
plt.xticks(rotation=90)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Group by neighbourhood and calculate the mean price, then sort in ascending order
df_price_cheapest = df.groupby(by='neighbourhood')['price'].mean().sort_values(ascending=True).reset_index()

# Plot the 10 cheapest neighbourhoods
plt.figure(figsize=(12, 6))
sns.barplot(x=df_price_cheapest['neighbourhood'].head(10), y=df_price_cheapest['price'].head(10), palette='coolwarm')

plt.title("Top 10 Cheapest Neighbourhoods", fontsize=16)
plt.xlabel("Neighbourhood", fontsize=12)
plt.ylabel("Average Price", fontsize=12)
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

Review score analysis

```python
# Group by neighbourhood and count the number of listings (assuming 'id' column exists)
neighbourhood_counts = df.groupby('neighbourhood')['id'].count().sort_values(ascending=False)

# Plot as a pie chart (top 10 neighbourhoods for better readability)
plt.figure(figsize=(10, 8))
neighbourhood_counts.head(10).plot.pie(autopct='%1.1f%%', startangle=140, shadow=True)

plt.title('Top 10 Neighbourhoods by Number of Listings')
plt.ylabel('')  # Hide y-axis label
plt.tight_layout()
plt.show()
```

```python
# Group by room_type and count the number of listings (using 'id')
room_type_counts = df.groupby('room_type')['id'].count().reset_index()

# Sort values for better display (optional)
room_type_counts = room_type_counts.sort_values(by='id', ascending=False)

# Plot the bar chart
plt.figure(figsize=(8, 6))
sns.barplot(x='room_type', y='id', data=room_type_counts, palette='Set2')

# Add labels and title
plt.title('Number of Listings per Room Type', fontsize=16)
plt.xlabel('Room Type', fontsize=12)
plt.ylabel('Number of Listings', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
